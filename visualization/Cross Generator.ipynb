```python
available_fips = county_data['RegionName'].unique()
available_fips.shape
```

```python
county_cross = county_cross[county_cross['FIPS'].isin(available_fips)].reset_index(drop=True)
county_cross.shape
```

```python
d = county_cross[['FIPS','CountyName','StateName']]
d.head(5)
```

```python
c = []
for i in d['StateName'].values:
    c.append(us_state_abbrev[i])
d['StateCode'] = c
d = d.sort_values(['StateName', 'CountyName']).reset_index(drop=True)
d.to_csv('UI/plots/cross.csv',index=False)
```
